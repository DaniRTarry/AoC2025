# Advent of Code Day 6

Puzzle text available at:
https://adventofcode.com/2025/day/6

```python
import os
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Day of calendar
day = 6
```

```python
# Set path to input
input_dir = Path(os.path.abspath('')).parent
filepath = os.path.join(input_dir,'inputs','input_day%02d.txt' %(day))
```

## Part 1

```python
problem_input_part1 = pd.read_csv(filepath,sep=r'\s+',header=None)
problem_input_part1
```

```python
# Separate numbers from operations
numbers = problem_input_part1.iloc[:-1].astype(int)
operations = problem_input_part1.iloc[-1]
```

Do the operation with the correct symbol and add the grand total

```python
# Insert row for operation result
ncols = numbers.columns.size
numbers_part1 = numbers.copy()
numbers_part1.loc[len(numbers)] = [0] * ncols

for i in range(ncols):
    operation = operations[i]
    vector = numbers.iloc[:,i].values
    if operation == '*':
        result = np.prod(vector)
    elif operation == '+':
        result = np.sum(vector)
    
    numbers_part1.iloc[-1,i] = result
```

```python
print(f'What is the grand total found by adding together all of the answers to the individual problems?:')
print(np.sum(numbers_part1.iloc[-1,:]))
```

## Part 2

```python
problem_input_part2 = []
with open(filepath, 'r') as f:
    for line in f:
        problem_input_part2.append(list(line.rstrip('\n')))
```

```python
right_to_left = np.flip(problem_input_part2,axis=1)
operations_str = right_to_left[-1,:]
operations_str = operations_str[operations_str != ' ']
numbers_part2 = right_to_left[:-1,:]
```

```python
# Init vars
vector = []
results = []
operation_idx = 0

# Loop through every column
for i in range(numbers_part2.shape[1]):
    col = ''.join(numbers_part2[:,i].tolist())
    
    # We do the operation once we got to a separator or reach the end of the list
    if col == '    ' or i == numbers_part2.shape[1] - 1:
        if i == numbers_part2.shape[1] - 1: # Append the last number of the file
            vector.append(int(col))
        
        if operations_str[operation_idx] == '*':
            result = np.prod(vector)
        elif operations_str[operation_idx] == '+':
            result = np.sum(vector)    

        results.append(result)
        vector = []
        operation_idx += 1
    else:
        vector.append(int(col))
```

```python
print(f"What is the grand total found by adding together all of the answers to the individual problems?")
print(int(np.sum(results)))
```
