# Advent of Code Day 5

Puzzle text available at:
https://adventofcode.com/2025/day/5

```python
import os
import numpy as np
from pathlib import Path
```

```python
# Day of calendar
day = 5
```

```python
# Set path to input
input_dir = Path(os.path.abspath('')).parent
```

```python
filepath = os.path.join(input_dir,'inputs','input_day%02d.txt' %(day))
with open(filepath, 'r') as f:
    problem_input = f.read().splitlines()
```

```python
# Split list between ranges and ingredients
split_idx = problem_input.index('')
ranges_list = problem_input[:split_idx]
ingredients_list = problem_input[split_idx + 1:]
print(ranges_list[:10])
print(ingredients_list[:10])
```

## Part 1

The first part of the file set the valid ranges. We have to identify which elements of the second half of the list are within the ranges.

```python
result_part1 = 0
for ingredient in ingredients_list:
    for ingredient_range in ranges_list:
        range_limits = ingredient_range.split('-')
        start, end = int(range_limits[0]), int(range_limits[1])
        current_range = range(start, end + 1)

        if int(ingredient) in current_range:
            result_part1 += 1
            break
```

```python
print(f'How many of the available ingredient IDs are fresh?: {int(result_part1)}')
```

## Part 2

How many valid ingredients range IDs there are?

```python
# Create all of the ranges
ranges = []
for range_str in ranges_list:
    if range_str:  # Skip empty strings
        parts = range_str.split('-')
        start, end = int(parts[0]), int(parts[1])
        ranges.append((start, end))

# Sort by start
ranges.sort()

# Merge overlapping ranges
merged = [ranges[0]]
for start2, end2 in ranges[1:]:
    start1 = merged[-1][0]
    end1 = merged[-1][1]
    
    if start2 <= end1 + 1:  #Overlapping or adjacent
        # Extend the range if the new end is larger
        merged[-1] = (start1, max(end1, end2))
    else:
        merged.append((start2, end2))

# Count total elements
result_part2 = sum(end - start + 1 for start, end in merged)
```

```python
print(f"How many ingredient IDs are considered to be fresh according to the fresh ingredient ID ranges?")
print(int(result_part2))
```
