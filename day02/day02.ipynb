# Advent of Code Day 2

```python
import os
import numpy as np
from pathlib import Path
```

```python
# Day of calendar
day = 2
```

```python
# Set path to input
input_dir = Path(os.path.abspath('')).parent
```

```python
filepath = os.path.join(input_dir,'inputs','input_day%02d.txt' %(day))
with open(filepath, 'r') as f:
    problem_input = f.readlines() # read the entire file at once and returns a list where each line is a separate string element
```

```python
IDs = []
# Split file elements by commas
lines = problem_input[0].strip().split(',')
for line in lines:
    # Split file IDs by dashes and convert to int
    IDs.append([int(x) for x in line.strip().split('-')])
IDs[:10]
```

## Part 1

Find the invalid IDs by looking for any ID which is made only of some sequence of digits repeated twice

```python
result_part1 = 0
for first_ID,last_ID in IDs:
    for id in range(first_ID,last_ID + 1):
        id_string = str(id)
        # Check if the ID can be divided into two equal sequences
        if len(id_string) % 2 == 0:
            # Check if the ID is made by a sequence of digiits repeatead twice
            if id_string[len(id_string)//2 :] == id_string[: len(id_string)//2]:
                result_part1 += id
```

```python
print(f"What do you get if you add up all of the invalid IDs?")
print(result_part1)
```

## Part 2

Now, an ID is invalid if it is made only of some sequence of digits repeated at least twice. So, 12341234 (1234 two times), 123123123 (123 three times), 1212121212 (12 five times), and 1111111 (1 seven times) are all invalid IDs

```python
result_part2 = 0
for first_ID, last_ID in IDs:
    for id in range(first_ID, last_ID + 1):
        id_string = str(id)
        n = len(id_string)
        for pattern_len in range(1, n // 2 + 1): #(min possible sequence, max possible sequence)
            # Check if the ID can be divided into two equal sequences
            if n % pattern_len == 0:
                pattern = id_string[:pattern_len]
                if pattern * (n // pattern_len) == id_string:
                    result_part2 += id
                    break  # Stop checking other patterns once found
```

```python
print(f"What do you get if you add up all of the invalid IDs using these new rules?")
print(result_part2)
```
